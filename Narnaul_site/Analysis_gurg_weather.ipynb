```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataframe
data = pd.read_csv('data/Weather Data/Data_gurg.csv', index_col = 0)
data.head()
```

```python
print(data.columns)
```

### Large amount of data
Therefore selecting every 10th minute

```python
data_red15= data.loc[::10, :]
data.index = pd.to_datetime(data.index)
```

### Analyzing the graphs of various features

```python
for col in ['Battery Level (V)','Relative Humidity (%)','Air Temperature (°C)','Global Radiation 1 (W/m2)','Solar Azimuth (°)']:
    plt.figure(figsize=[30,2])
    temp = data[data.index.hour == 14][col]
    
    # Comparing relative humidity only at 1:00 pm
    plt.plot(temp)
    plt.title(col)
plt.show()
```
